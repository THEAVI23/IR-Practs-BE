```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
data.head()
```

```python
data.columns
```

```python
data.dtypes
```

```python
X = data.iloc[:, :2]
```

```python
model = AgglomerativeClustering(n_clusters=3) 
clusters = model.fit_predict(X)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=data, palette='viridis')
plt.title('Agglomerative Hierarchical Clustering of Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
linked = linkage(X, method='ward')  # 'ward' minimizes variance within clusters
dendrogram(linked, orientation='top', labels=data.index, distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Agglomerative Hierarchical Clustering')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
test_variable = np.array([[5.5, 2.4]]) 
```

```python
test_cluster = model.fit_predict(np.vstack([X, test_variable]))[-1]
```

```python
print(f"Test variable: Sepal Length = {test_variable[0][0]}, Sepal Width = {test_variable[0][1]}")
print(f"The test variable belongs to Cluster: {test_cluster}")
```
